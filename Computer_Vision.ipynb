###Drawing shapes

````python
image = np.zeros((512,512,3))
````

````python
cv2_imshow(image)
````

````python
#Rectangle
rec_img=cv2.rectangle(image,pt1=(200,200),pt2=(100,100),color=(255,0,0),thickness=1)
````

````python
cv2_imshow(rec_img)
````

````python
#Circle
circle=cv2.circle(image,center=(100,400),radius=50,color=(0,0,255),thickness=-1)
````

````python
cv2_imshow(circle)
````

````python
#line
line=cv2.line(image,pt1=(0,0),pt2=(512,512),color=(0,100,0),thickness=2)
cv2_imshow(line)
````

````python
#Text
line=cv2.putText(image,org=(200,400),fontScale=4,color=(0,255,255),thickness=2,lineType=cv2.LINE_AA,text="Hello",fontFace=cv2.FONT_ITALIC)
cv2_imshow(line)
````

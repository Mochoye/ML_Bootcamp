# Computer Vision Project

### Input Image and Resize It



```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/dp new.jpeg')
```

```python
cv2_imshow(image)
```

```python
image_resized = cv2.resize(image,None,fx=0.5,fy=0.5)
```

```python
cv2_imshow(image_resized)
```

### Clear the Impurities

The impurities may be colour variations due to which filters may lag

```python
image_cleared = cv2.medianBlur(image_resized,3)
image_cleared = cv2.medianBlur(image_resized,3)
image_cleared = cv2.medianBlur(image_resized,3)

image_cleared = cv2.edgePreservingFilter(image_cleared, sigma_s=5)
```

```python
cv2_imshow(image_cleared)
```

```python
cv2_imshow(image_cleared)
```

### Apply Filtering

```python
#Bilateral Image Filtering

image_filtered = cv2.bilateralFilter(image_cleared,3,10,5)

for i in range(2):
  image_filtered = cv2.bilateralFilter(image_filtered,3,20,10)

for i in range(3):
  image_filtered = cv2.bilateralFilter(image_filtered,5,30,10)
```

```python
cv2_imshow(image_filtered)
```

### Tune the Art

Sharpening / dehazing / color enhancement

Sharpening : Main Image - Blur of that image

add_weighted =img1 x alpha + img2 * beta +gamma

```python
#sharpening the image using addWeighted()

guassian_mask = cv2.GaussianBlur(image_filtered,(7,7),2)

image_sharp = cv2.addWeighted(image_filtered,1.5,guassian_mask,-0.5,0)
image_sharp = cv2.addWeighted(image_filtered,1.4,guassian_mask,-0.2,10)
```

```python
cv2_imshow(image_sharp)
```
